```python
import re
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.infocasas.com.uy/venta'
```

```python
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
soup
```

###Propieadades

```python
container = soup.find_all('div', class_='lc-dataWrapper')
container
```

```python
len(container)
```

```python
properties = []

for post in container:
    # Scrapeo todos los valores
    inmobiliaria = post.find('div', class_='lc-owner-name')
    foo = post.find('strong', class_='lc-location body body-1 body-bold medium')
    bat = foo.get_text().split(',')
    bar = bat[0].split(' en ')
    bar_price  = post.find('div', class_='lc-price').get_text().split('+')
    bar_price = bar_price[0]
    bat_bed = post.find('div', class_='lc-typologyTag').get_text()
    bat_bed = bat_bed[0]
    bat_size = post.find('div', class_='lc-typologyTag').get_text()
   # print(bat_size)
    city = bat[1].strip()
    town = bar[1]
    # Genero el diccionario
    property = {'inmobiliaria': inmobiliaria.get_text(),
                'ciudad': city,
                'Barrio': town,
                'Precio': bar_price,
                'Dormitorios': bat_bed}
    properties.append(property)

properties
```

```python
import json
with open('../titles.json', 'w') as outfile:
    json.dump(properties,outfile, indent=4)
```
